```python
## Task 2
class Document:
  def __init__(self, name, exchange):
    self.name = name
    self.exchange = exchange

  def display_info(self):
    print(f"Название валюты: {self.name} \nОбменный курс: {self.exchange}")

class CurrencyRate(Document):
  def __init__(self, name, code, exchange):
    super().__init__(name, exchange)
    self.code = code

  def update_rate(self, new_rate):
    self.exchange = new_rate
  def rate_info(self):
    self.display_info()
    print(f"Код валюты: {self.code}")

y = CurrencyRate("Pickme", 11037, 42)
y.update_rate(52)
y.rate_info()
```

```python
## Task 4
class Publisher:
    def __init__(self, name):
        self.name = name
        self.magazines = []

    def create_magazine(self, name):
        magazine = Magazine(name, self)
        self.magazines.append(magazine)
        return magazine

    def __repr__(self):
        return f"Издательство(название={self.name}, журналы={self.magazines})"

class Magazine:
    def __init__(self, name, publisher):
        self.name = name
        self.publisher = publisher

    def __repr__(self):
        return f"Журнал(название={self.name}, издательство={self.publisher.name})"

publisher = Publisher("Издательство")
magazine1 = publisher.create_magazine("Журнал 1")
magazine2 = publisher.create_magazine("Журнал 2")

print(publisher)
print(magazine1)
print(magazine2)
```

```python
## Task 5
class MusicGroup:
    def __init__(self, name, genre):
        self.name = name
        self.genre = genre

class Concert:
    def __init__(self):
        self.groups = []

    def add_group(self, group):
        self.groups.append(group)

    def print_groups(self):
        for group in self.groups:
            print(f"Название: {group.name}, Жанр: {group.genre}")

group1 = MusicGroup("Venom", "rock")
group2 = MusicGroup("Pickme", "pop")

concert = Concert()

concert.add_group(group1)
concert.add_group(group2)

concert.print_groups()
```

```python
## Task 6
from abc import ABC, abstractmethod

class Room(ABC):
    @abstractmethod
    def calculateArea(self):
        pass

    @abstractmethod
    def paint(self):
        pass

class LivingRoom(Room):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculateArea(self):
        return self.length * self.width

    def paint(self):
        print("Покраска гостиной")

class Bedroom(Room):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculateArea(self):
        return self.length * self.width

    def paint(self):
        print("Покраска спальни")

living_room = LivingRoom(5, 4)
bedroom = Bedroom(3, 3)

print(f"Площадь гостиной: {living_room.calculateArea()} кв.м.")
print(f"Площадь спальни: {bedroom.calculateArea()} кв.м.")

living_room.paint()
bedroom.paint()
```

```python
## Task 7
from abc import ABC, abstractmethod

class ICurrencyRate(ABC):
    @abstractmethod
    def updateExchangeRate(self, rate):
        pass

    @abstractmethod
    def getRateInfo(self):
        pass

class CurrencyRate(ICurrencyRate):
    def __init__(self, currencyName, currencyCode, exchangeRate):
        self.currencyName = currencyName
        self.currencyCode = currencyCode
        self.exchangeRate = exchangeRate

    def updateExchangeRate(self, rate):
        self.exchangeRate = rate

    def getRateInfo(self):
        return (f"Название валюты: {self.currencyName}, "
                f"Код валюты: {self.currencyCode}, "
                f"Курс обмена: {self.exchangeRate}")

usd_rate = CurrencyRate("Доллар США", "USD", 75.5)
print(usd_rate.getRateInfo())

usd_rate.updateExchangeRate(76.0)
print(usd_rate.getRateInfo())
```

```python
## Task 8
from abc import ABC, abstractmethod

class CurrencyRate(ABC):
    def __init__(self, currencyName, currencyCode, exchangeRate):
        self.currencyName = currencyName
        self.currencyCode = currencyCode
        self.exchangeRate = exchangeRate

    @abstractmethod
    def convertCurrency(self, amount):
        pass

    def updateExchangeRate(self, rate):
        self.exchangeRate = rate

    def getRateInfo(self):
        return (f"Название валюты: {self.currencyName}, "
                f"Код валюты: {self.currencyCode}, "
                f"Курс обмена: {self.exchangeRate}")

class CryptoCurrencyRate(CurrencyRate):
    def convertCurrency(self, amount):
        return amount * self.exchangeRate * 0.99  # Учет комиссии

class FiatCurrencyRate(CurrencyRate):
    def convertCurrency(self, amount):
        return amount * self.exchangeRate

bitcoin_rate = CryptoCurrencyRate("Bitcoin", "BTC", 50000)
print(bitcoin_rate.getRateInfo())
print(f"Конвертация 2 BTC: {bitcoin_rate.convertCurrency(2)} USD")

usd_rate = FiatCurrencyRate("Доллар США", "USD", 75.5)
print(usd_rate.getRateInfo())
print(f"Конвертация 1000 USD: {usd_rate.convertCurrency(1000)} RUB")
```
